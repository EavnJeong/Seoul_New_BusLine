```python
from sklearn.preprocessing import minmax_scale

links['X'] = (links['SourceX'] + links['TargetX'])/2
links['Y'] = (links['SourceY'] + links['TargetY'])/2

max(links['X']), min(links['X']), max(links['Y']), min(links['Y']) , max(links['X'])- min(links['X']), max(links['Y'])-min(links['Y'])
links['X']=minmax_scale(links['X'], feature_range=(0, 1))
links['Y']=minmax_scale(links['Y'], feature_range=(0, 1))
links['XY']=(links['SourceX']-links['TargetX'])**2+(links['SourceY']-links['TargetY'])**2
links
```
